# 运算符号

数学中有各种运算,如:加减乘除等,而运算就要用到运算符,如: +, -, *, ÷ 等.

在c语言中同样有运算符,只是写法可能有所不同.在c语言中的运算就是对数据进行操作,处理的过程.那么运算符就是指定该运算的处理方式.

c语言中的运算符有:
+ 算术运算符
+ 赋值运算符
+ 关系运算符
+ 逻辑运算符
+ 三目运算符

# 算术运算符

c语言基本算术运算符如下:
+ 加法: +
+ 减法: -
+ 乘法: *
+ 除法: /
+ 求余: %
+ 自增: ++
+ 自减: --

除法运算中需注意:

如果相除的**两个数都是整数**的话.则**结果也为整数**,小数部分省略.如8/3=2;而两数中**有一个为小数,则结果为小数**,如:9.0/2=4.500000.

取余运算中注意:

该运算只适合用两个整数进行取余运算,如:10%3=1;而10.0%3则是错误的;运算后的的符号取决于被模数的符号,如(-10)%3=-1;而10%(-3)=1 .

注:c运算中没有乘方这个运算符,也不能用 * , ÷ 等算术符号.


### 自增与自减运算符

自增运算符为"++",其功能是使变量的值自增1;自减运算符为"--",其功能是使变量自减1;它们经常使用在循环中.自增自减运算符有以下几种形式:
```
++a :a自增1后,再取值,--a :a自减1后,再取值;
a++ :a取值后,a的值再自增1, a-- :a取值后,a的值再自减1;
```

注意:无论是a++还是++a都等同于a=a+1,在表达式执行完毕后a的值都自增了1,无论是a--还是--a都等同于a=a-1,在表达式执行完毕后a的值都自减少1。

以下代码为例:

```
#include <stdio.h>
int main()
{
    int a=100;
    printf("%d\n",a++);
    printf("%d\n",++a);
    printf("%d\n",--a);
    printf("%d\n",a--);
    printf("%d\n",a+1);  //注意:这里并没有改变a的值,输出的就是a+1的值
    printf("%d\n",a);
    return 0;
}
```
运行结果为:
```
100
102
101
101
101
100
```

一个例子:
```
int a = 3;b = 4;
a+b%(a+b)/a+a*b-b
```
运算结果为:
```
12
```

# 赋值运算符

c语言中的赋值运算符分为:**简单赋值运算符**和**复合赋值运算符**;

复合赋值运算符就是在简单赋值符"="之前加上其他运算符构成,例如: "+=","-=","/=","%=";

例如:
```
int a=3;
a+=5;
```
分析:定义整型变量a并赋值为3, a+=5;这个算式就等价于a = a+5;

同理:a*=5;等价于a=a * 5;

# 关系运算符

+ > 大于;
+ >= 大于等于;
+ < 小于;
+ <= 小于等于;
+ == 等于;
+ != 不等于;

关系表达式的值是"真"和"假",在c程序用整数1和0表示.如:
```
int a = 10;
int b = 4;
printf("%d\n", a>b);  //10>4,为真,输出1
printf("%d\n", a==b);  //为假,输出0
```

# 逻辑运算符(logical operator)

c语言的逻辑运算符有:
+ && :and 逻辑与;
+ || :or 逻辑或;
+ ! :not 逻辑非;


# 三目运算符

如果...就...,这样的语句在c中可以使用三目运算符来表示,具体的格式为:
```
表达式1 ? 表达式2 : 表达式3;
```
语句的执行过程为:
先判断表达式1是否为真,如果是真就执行表达式2;如果为假就执行表达式3;

例码:
```
#include <stdio.h>
int main()
{
    double money = 12;
    double cost = 11.5;
    printf("钱够不够呢?\n");
    printf(money>=cost?"yes":"no");
    
    return 0;
}
```

5. < , <= , > , >=
6. == , !=
7. &&
8. ||
9. ?: (三目运算符)
